Given two linked list of the same size, the task is to create a new linked list using those linked lists. The condition is that the greater node among both linked list will be added to the new linked list.

**Examples:**
Input: list1 = 5->2->3->8
list2 = 1->7->4->5
Output: New list = 5->7->4->8

Input:list1 = 2->8->9->3
list2 = 5->3->6->4
Output: New list = 5->8->9->4
</aside>

```python
# Definition for a singly linked list node
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

# Function to create a new linked list using two input linked lists
def create_new_linked_list(list1, list2):
    # Initialize a dummy node as the head of the new linked list
    dummy = Node(0)
    # Pointer to the current node of the new linked list
    curr = dummy

    # Traverse both input linked lists simultaneously
    while list1 is not None and list2 is not None:
        # Compare the values of the current nodes from both lists
        if list1.data >= list2.data:
            # Add the greater value to the new linked list
            curr.next = Node(list1.data)
            list1 = list1.next
        else:
            curr.next = Node(list2.data)
            list2 = list2.next
        # Move the pointer of the new linked list to the next node
        curr = curr.next

    # Add the remaining nodes of list1, if any
    while list1 is not None:
        curr.next = Node(list1.data)
        list1 = list1.next
        curr = curr.next

    # Add the remaining nodes of list2, if any
    while list2 is not None:
        curr.next = Node(list2.data)
        list2 = list2.next
        curr = curr.next

    # Return the new linked list (excluding the dummy head)
    return dummy.next

# Create the first linked list: 5->2->3->8
list1 = Node(5)
list1.next = Node(2)
list1.next.next = Node(3)
list1.next.next.next = Node(8)

# Create the second linked list: 1->7->4->5
list2 = Node(1)
list2.next = Node(7)
list2.next.next = Node(4)
list2.next.next.next = Node(5)

# Create the new linked list using the above two lists
new_list = create_new_linked_list(list1, list2)

# Print the new linked list: 5->7->4->8
while new_list is not None:
    print(new_list.data, end="->")
    new_list = new_list.next
```

ðŸ’¡ **Question 2**

Write a function that takes a list sorted in non-decreasing order and deletes any duplicate nodes from the list. The list should only be traversed once.

For example if the linked list is 11->11->11->21->43->43->60 then removeDuplicates() should convert the list to 11->21->43->60.

**Example 1:**
    Input:
LinkedList: 
11->11->11->21->43->43->60
Output:
11->21->43->60

</aside>

```python
# Definition for a singly linked list node
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

# Function to remove duplicate nodes from a sorted linked list
def remove_duplicates(head):
    # Check if the linked list is empty or has only one node
    if head is None or head.next is None:
        return head

    # Initialize a pointer to traverse the linked list
    current = head

    # Traverse the linked list
    while current.next is not None:
        # Compare the value of the current node with the value of the next node
        if current.data == current.next.data:
            # Move the pointer of the next node until the values are not equal or the end of the list is reached
            current.next = current.next.next
        else:
            # Move to the next node with a different value
            current = current.next

    # Return the modified linked list
    return head
# Create the linked list: 11->11->11->21->43->43->60
head = Node(11)
head.next = Node(11)
head.next.next = Node(11)
head.next.next.next = Node(21)
head.next.next.next.next = Node(43)
head.next.next.next.next.next = Node(43)
head.next.next.next.next.next.next = Node(60)

# Remove duplicate nodes from the linked list
head = remove_duplicates(head)

# Print the modified linked list: 11->21->43->60
current = head
while current is not None:
    print(current.data, end="->")
    current = current.next
```

Given a linked list of sizeÂ **N**. The task is to reverse everyÂ **k**Â nodes (where k is an input to the function) in the linked list. If the number of nodes is not a multiple ofÂ *k*Â then left-out nodes, in the end, should be considered as a group and must beÂ reversed (See Example 2 for clarification).

**Example 1:**
    Input:
LinkedList: 1->2->2->4->5->6->7->8
K = 4
Output:4Â 2Â 2Â 1Â 8Â 7Â 6Â 5
Explanation:
The first 4 elements 1,2,2,4 are reversed first
and then the next 4 elements 5,6,7,8. Hence, the
resultant linked list is 4->2->2->1->8->7->6->5.

</aside>

```python
# Definition for a singly linked list node
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

# Function to reverse every k nodes in a linked list
def reverse_k_nodes(head, k):
    # Check if the linked list is empty or has only one node
    if head is None or head.next is None:
        return head

    # Initialize pointers
    current = head
    prev = None
    next_node = None

    # Reverse the first k nodes
    count = 0
    while current is not None and count < k:
        next_node = current.next
        current.next = prev
        prev = current
        current = next_node
        count += 1

    # Recursively reverse the remaining nodes and connect them to the reversed portion
    if next_node is not None:
        head.next = reverse_k_nodes(next_node, k)

    # Return the new head of the reversed linked list
    return prev


# Create the linked list: 1->2->2->4->5->6->7->8
head = Node(1)
head.next = Node(2)
head.next.next = Node(2)
head.next.next.next = Node(4)
head.next.next.next.next = Node(5)
head.next.next.next.next.next = Node(6)
head.next.next.next.next.next.next = Node(7)
head.next.next.next.next.next.next.next = Node(8)

# Reverse every k nodes in the linked list
k = 4
head = reverse_k_nodes(head, k)

# Print the reversed linked list
current = head
while current is not None:
    print(current.data, end="->")
    current = current.next
```

ðŸ’¡ **Question 4**

Given a linked list, write a function to reverse every alternate k nodes (where k is an input to the function) in an efficient way. Give the complexity of your algorithm.

**Example:**
    Inputs:   1->2->3->4->5->6->7->8->9->NULL and k = 3
Output:   3->2->1->4->5->6->9->8->7->NULL.

</aside>

```python
# Definition for a singly linked list node
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

# Function to reverse every alternate k nodes in a linked list
def reverse_alternate_k_nodes(head, k):
    # Check if the linked list is empty or has only one node
    if head is None or head.next is None:
        return head

    # Initialize pointers
    current = head
    prev = None
    next_node = None

    # Reverse the first k nodes
    count = 0
    while current is not None and count < k:
        next_node = current.next
        current.next = prev
        prev = current
        current = next_node
        count += 1

    # Connect the last node of the reversed group to the first node of the next group
    if next_node is not None:
        head.next = next_node

    # Skip the next k nodes and recursively reverse the remaining alternate groups
    count = 0
    while count < k-1 and next_node is not None:
        next_node = next_node.next
        count += 1

    if next_node is not None:
        next_node.next = reverse_alternate_k_nodes(next_node.next, k)

    # Return the head of the modified linked list
    return prev


head = Node(1)
head.next = Node(2)
head.next.next = Node(3)
head.next.next.next = Node(4)
head.next.next.next.next = Node(5)
head.next.next.next.next.next = Node(6)
head.next.next.next.next.next.next = Node(7)
head.next.next.next.next.next.next.next = Node(8)
head.next.next.next.next.next.next.next.next = Node(9)

# Reverse every alternate k nodes in the linked list
k = 3
head = reverse_alternate_k_nodes(head, k)

# Print the modified linked list
current = head
while current is not None:
    print(current.data, end="->")
    current = current.next
```

ðŸ’¡ **Question 5**

Given a linked list and a key to be deleted. Delete last occurrence of key from linked. The list may have duplicates.

**Examples**:
    Input:   1->2->3->5->2->10, key = 2
Output:  1->2->3->5->10

</aside>

```python
# Definition for a singly linked list node
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

# Function to delete the last occurrence of a key from a linked list
def delete_last_occurrence(head, key):
    # Check if the linked list is empty
    if head is None:
        return head

    # Initialize pointers
    prev = None
    curr = head
    last = None
    last_prev = None

    # Traverse the linked list
    while curr is not None:
        if curr.data == key:
            last = curr
            last_prev = prev
        prev = curr
        curr = curr.next

    # Delete the last occurrence if found
    if last is not None:
        if last_prev is None:
            # If the last occurrence is the head, update the head
            head = head.next
        else:
            # Set the next pointer of the previous node to the next node of the last occurrence
            last_prev.next = last.next

    # Return the head of the modified linked list
    return head

# Create the linked list: 1->2->3->5->2->10
head = Node(1)
head.next = Node(2)
head.next.next = Node(3)
head.next.next.next = Node(5)
head.next.next.next.next = Node(2)
head.next.next.next.next.next = Node(10)

# Delete the last occurrence of key = 2 from the linked list
key = 2
head = delete_last_occurrence(head, key)

# Print the modified linked list
current = head
while current is not None:
    print(current.data, end="->")
    current = current.next
```

Given two sorted linked lists consisting ofÂ **N**Â andÂ **M**Â nodes respectively. The task is to merge both of the lists (in place) and return the head of the merged list.

**Examples:**

Input:Â a: 5->10->15,Â b: 2->3->20

Output:Â 2->3->5->10->15->20

Input:Â a: 1->1, b: 2->4

Output:Â 1->1->2->4

</aside>

```python
# Definition for a singly linked list node
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

# Function to merge two sorted linked lists in-place
def merge_sorted_lists(a, b):
    # Create a dummy node as the head of the merged list
    dummy = Node(0)
    result = dummy

    # Traverse both lists and merge them
    while a is not None and b is not None:
        if a.data <= b.data:
            result.next = a
            a = a.next
        else:
            result.next = b
            b = b.next
        result = result.next

    # Connect the remaining nodes of the non-empty list
    if a is not None:
        result.next = a
    elif b is not None:
        result.next = b

    # Return the head of the merged list
    return dummy.next
# Create the first sorted linked list: 5->10->15
a = Node(5)
a.next = Node(10)
a.next.next = Node(15)

# Create the second sorted linked list: 2->3->20
b = Node(2)
b.next = Node(3)
b.next.next = Node(20)

# Merge the two sorted linked lists
merged_head = merge_sorted_lists(a, b)

# Print the merged linked list
current = merged_head
while current is not None:
    print(current.data, end="->")
    current = current.next
```

Given a **Doubly Linked List**, the task is to reverse the given Doubly Linked List.

```python
# Definition for a doubly linked list node
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None

# Function to reverse a doubly linked list
def reverse_doubly_linked_list(head):
    # Check if the doubly linked list is empty or has only one node
    if head is None or head.next is None:
        return head

    # Initialize pointers
    current = head
    prev = None

    # Traverse the doubly linked list and reverse each node
    while current is not None:
        next_node = current.next
        current.next = prev
        current.prev = next_node
        prev = current
        current = next_node

    # Return the head of the reversed doubly linked list
    return prev

# Create the doubly linked list: 1<->2<->3<->4<->5
head = Node(1)
node2 = Node(2)
node3 = Node(3)
node4 = Node(4)
node5 = Node(5)
head.next = node2
node2.prev = head
node2.next = node3
node3.prev = node2
node3.next = node4
node4.prev = node3
node4.next = node5
node5.prev = node4

# Reverse the doubly linked list
reversed_head = reverse_doubly_linked_list(head)

# Print the reversed doubly linked list forwards
current = reversed_head
while current is not None:
    print(current.data, end="<->")
    current = current.next



# Print the reversed doubly linked list backwards
current = reversed_head
while current.next is not None:
    current = current.next

while current is not None:
    print(current.data, end="<->")
    current = current.prev
```

Given a doubly linked list and a position. The task is to delete a node from given position inÂ a doubly linked list.
Input:
LinkedList = 1 <--> 3 <--> 4
x = 3
Output:1 3
Explanation:After deleting the node at
position 3 (position starts from 1),
the linked list will be now as 1->3.

</aside>

```python
# Definition for a doubly linked list node
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None

# Function to delete a node from a doubly linked list at a given position
def delete_node_at_position(head, position):
    # Check if the doubly linked list is empty
    if head is None:
        return head

    # Case 1: Delete the head node
    if position == 1:
        head = head.next
        if head is not None:
            head.prev = None
        return head

    # Traverse the doubly linked list to the node at the given position
    current = head
    count = 1
    while current is not None and count < position:
        current = current.next
        count += 1

    # If the position exceeds the length of the linked list, return the original list
    if current is None:
        return head

    # Update the previous node's next pointer and the next node's prev pointer
    current.prev.next = current.next
    if current.next is not None:
        current.next.prev = current.prev

    # Delete the current node
    current.prev = None
    current.next = None

    # Return the head of the modified doubly linked list
    return head

# Create the doubly linked list: 1<->3<->4
head = Node(1)
node2 = Node(3)
node3 = Node(4)
head.next = node2
node2.prev = head
node2.next = node3
node3.prev = node2

# Delete the node at position 3 from the doubly linked list
position = 3
head = delete_node_at_position(head, position)

# Print the modified doubly linked list forwards
current = head
while current is not None:
    print(current.data, end="<->")
    current = current.next



# Print the modified doubly linked list backwards
current = head
while current.next is not None:
    current = current.next

while current is not None:
    print(current.data, end="<->")
    current = current.prev
```
