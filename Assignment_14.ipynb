Given a linked list ofÂ **N**Â nodes such that it may contain a loop.

A loop here means that the last node of the link list is connected to the node at position X(1-based index). If the link list does not have any loop, X=0.

Remove the loop from the linked list, if it is present, i.e. unlink the last node which is forming the loop.

**Example 1:**
    Input:
N = 3
value[] = {1,3,4}
X = 2
Output:1
Explanation:The link list looks like
1 -> 3 -> 4
     ^    |
     |____|
A loop is present. If you remove it
successfully, the answer will be 1.


</aside>

```python
def remove_loop(head):
    slow = fast = head

    # Step 1: Find the meeting point of slow and fast pointers
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        if slow == fast:
            break

    # Step 2: No loop found, return the original linked list
    if fast is None or fast.next is None:
        return head

    # Step 3: Move slow pointer back to the head
    slow = head

    # Step 4: Move slow and fast pointers to find the node just before the loop
    while slow.next != fast.next:
        slow = slow.next
        fast = fast.next

    # Step 5: Remove the loop by setting the next pointer to null
    fast.next = None

    return head


class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

# Create the linked list
head = Node(1)
node2 = Node(3)
node3 = Node(4)

head.next = node2
node2.next = node3
node3.next = node2  # Create a loop by connecting the last node to the second node

# Call the remove_loop function
head = remove_loop(head)

# Print the resulting linked list
current = head
while current:
    print(current.value, end=' ')
    current = current.next
```

ðŸ’¡ **Question 2**

A numberÂ **N**Â is represented in Linked List such that each digit corresponds to a node in linked list. You need to add 1 to it.

**Example 1:**
    Input:
LinkedList: 4->5->6
Output:457

Example 2:
Input:
LinkedList: 1->2->3
Output:124

</aside>

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

def add_one(head):
    # Step 1: Reverse the linked list
    prev = None
    current = head
    while current:
        next_node = current.next
        current.next = prev
        prev = current
        current = next_node
    head = prev

    # Step 2: Add 1 to the linked list
    carry = 1
    current = head
    while current:
        # Add carry to the current node's value
        current.value += carry

        # Update carry and node value to ensure single digit
        carry = current.value // 10
        current.value %= 10

        # Move to the next node
        prev = current
        current = current.next

    # Step 3: Add a new node with carry if it exists
    if carry:
        new_node = Node(carry)
        prev.next = new_node

    # Step 4: Reverse the linked list again
    prev = None
    current = head
    while current:
        next_node = current.next
        current.next = prev
        prev = current
        current = next_node
    head = prev

    return head

# Create the linked list: 4 -> 5 -> 6
head = Node(4)
node2 = Node(5)
node3 = Node(6)

head.next = node2
node2.next = node3

# Call the add_one function
head = add_one(head)

# Print the resulting linked list
current = head
while current:
    print(current.value, end='')
    current = current.next

```

Given a Linked List of size N, where every node represents a sub-linked-list and contains two pointers:(i) aÂ **next**Â pointer to the next node,(ii) aÂ **bottom**Â pointerÂ to a linked list where this node is head.Each of theÂ sub-linked-list is in sorted order.Flatten the Link List such that all the nodes appear in a single level while maintaining the sorted order.Â **Note:**Â The flattened list will be printed using the bottom pointer instead of next pointer.

**Example 1:**
    Input:
5 -> 10 -> 19 -> 28
|     |     |     |
7     20    22   35
|           |     |
8          50    40
|                 |
30               45
Output:Â 5-> 7-> 8- > 10 -> 19-> 20->
22-> 28-> 30-> 35-> 40-> 45-> 50.
Explanation:
The resultant linked lists has every
node in a single level.(Note:| represents the bottom pointer.)

Example 2:
    
    Input:
5 -> 10 -> 19 -> 28
|          |
7          22
|          |
8          50
|
30
Output: 5->7->8->10->19->22->28->30->50
Explanation:
The resultant linked lists has every
node in a single level.

(Note:| represents the bottom pointer.)
</aside>

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None
        self.bottom = None

def merge_lists(list1, list2):
    # Base cases: if either of the lists is empty
    if not list1:
        return list2
    if not list2:
        return list1

    # Compare the heads of the two lists and select the smaller one
    if list1.value <= list2.value:
        merged_list = list1
        merged_list.bottom = merge_lists(list1.bottom, list2)
    else:
        merged_list = list2
        merged_list.bottom = merge_lists(list1, list2.bottom)

    return merged_list

def flatten_list(head):
    # Base case: empty list or single node
    if head is None or head.next is None:
        return head

    # Recursively flatten the sub-linked list starting from the next node
    flattened_list = flatten_list(head.next)

    # Merge the current sub-linked list with the flattened sub-linked list
    head = merge_lists(head, flattened_list)

    return head

def print_list(head):
    current = head
    while current:
        print(current.value, end=' ')
        current = current.bottom

    print()

# Example 1
head1 = Node(5)
head1.next = Node(10)
head1.next.next = Node(19)
head1.next.next.next = Node(28)

head1.bottom = Node(7)
head1.next.bottom = Node(20)
head1.next.next.bottom = Node(22)
head1.next.next.next.bottom = Node(35)

head1.next.bottom.bottom = Node(8)
head1.next.next.bottom.bottom = Node(50)

head1.next.next.next.bottom = Node(30)
head1.next.next.next.bottom.bottom = Node(45)

print("Flattened list - Example 1:")
head1 = flatten_list(head1)
print_list(head1)

# Example 2
head2 = Node(5)
head2.next = Node(10)
head2.next.next = Node(19)
head2.next.next.next = Node(28)

head2.bottom = Node(7)
head2.next.bottom = Node(22)

head2.next.bottom.bottom = Node(8)
head2.next.next.bottom = Node(50)

head2.next.next.next.bottom = Node(30)

print("Flattened list - Example 2:")
head2 = flatten_list(head2)
print_list(head2)
```

You are given a specialÂ linked listÂ withÂ **N**Â nodes where each node has aÂ next pointer pointing to itsÂ next node. You are also givenÂ **M**Â random pointers, where you will be givenÂ **M**Â number of pairs denoting two nodesÂ **a**Â andÂ **b**Â Â **i.e.Â a->arb = b**Â (arb is pointer to random node)**.**

Construct a copy of the given list. The copy should consist of exactlyÂ **N**Â new nodes, where each new node has its value set to the value of its corresponding original node. Both the next and random pointer of the new nodes should point to new nodes in the copied list such that the pointers in the original list and copied list represent the same list state. None of the pointers in the new list should point to nodes in the original list.

For example, if there are two nodesÂ **X**Â andÂ **Y**Â in the original list, whereÂ **X.arb**Â **-->**Â **Y**, then for the corresponding two nodesÂ **x**Â andÂ **y**Â in the copied list,Â **x.arb --> y.**

Return the head of the copied linked list.

Input:
N = 4, M = 2
value = {1,2,3,4}
pairs = {{1,2},{2,4}}
Output:1
Explanation:In this test case, there
are 4 nodes in linked list.Â  Among these
4 nodes,Â  2 nodes have arbitrary pointer
set, rest two nodes have arbitrary pointer
as NULL. Second line tells us the value
of four nodes. The third line gives the
information about arbitrary pointers.
The first node arbitrary pointer is set to
node 2.Â  The second nodeÂ arbitrary pointer
is set to node 4.

Example 2:

Input:
N = 4, M = 2
value[] = {1,3,5,9}
pairs[] = {{1,1},{3,4}}
Output:1
Explanation:In the given testcase ,
applying the method as stated in the
above example, the output will be 1.

</aside>

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None
        self.random = None

def copyRandomList(head):
    if head is None:
        return None

    # Step 1: Create a hash map to store the mapping between original and copied nodes
    node_map = {}

    # Step 2: Create a copy of each node and store the mapping in the hash map
    current = head
    while current:
        copied_node = Node(current.value)
        node_map[current] = copied_node
        current = current.next

    # Step 3: Traverse the original list again and set the next and random pointers of copied nodes
    current = head
    while current:
        copied_node = node_map[current]
        copied_node.next = node_map.get(current.next)
        copied_node.random = node_map.get(current.random)
        current = current.next

    # Step 4: Return the head of the copied linked list
    return node_map[head]

# Example 1
head1 = Node(1)
head1.next = Node(2)
head1.next.next = Node(3)
head1.next.next.next = Node(4)

head1.random = head1.next
head1.next.random = head1.next.next.next

print("Copied linked list - Example 1:")
copied_head1 = copyRandomList(head1)
current = copied_head1
while current:
    print(f"Node value: {current.value}")
    if current.random:
        print(f"Random pointer: {current.random.value}")
    else:
        print("Random pointer: None")
    current = current.next
    print()

# Example 2
head2 = Node(1)
head2.next = Node(3)
head2.next.next = Node(5)
head2.next.next.next = Node(9)

head2.random = head2
head2.next.next.random = head2.next.next.next

print("Copied linked list - Example 2:")
copied_head2 = copyRandomList(head2)
current = copied_head2
while current:
    print(f"Node value: {current.value}")
    if current.random:
        print(f"Random pointer: {current.random.value}")
    else:
        print("Random pointer: None")
    current = current.next
    print()
```

Question 5

Given theÂ `head`Â of a singly linked list, group all the nodes with odd indices together followed by the nodes with even indices, and returnÂ *the reordered list*.

TheÂ **first**Â node is consideredÂ **odd**, and theÂ **second**Â node isÂ **even**, and so on.

Note that the relative order inside both the even and odd groups should remain as it was in the input.

You must solve the problemÂ inÂ `O(1)`Â extra space complexity andÂ `O(n)`Â time complexity.

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def oddEvenList(head):
    if not head or not head.next:
        return head

    odd_ptr = head
    even_ptr = head.next
    even_head = even_ptr

    while even_ptr and even_ptr.next:
        odd_ptr.next = even_ptr.next
        odd_ptr = odd_ptr.next
        even_ptr.next = odd_ptr.next
        even_ptr = even_ptr.next

    odd_ptr.next = even_head

    return head


# Example 1
head1 = ListNode(1)
head1.next = ListNode(2)
head1.next.next = ListNode(3)
head1.next.next.next = ListNode(4)
head1.next.next.next.next = ListNode(5)

print("Original list - Example 1:")
current = head1
while current:
    print(current.val, end=" ")
    current = current.next
print()

reordered_head1 = oddEvenList(head1)

print("Reordered list - Example 1:")
current = reordered_head1
while current:
    print(current.val, end=" ")
    current = current.next
print()

# Example 2
head2 = ListNode(2)
head2.next = ListNode(1)
head2.next.next = ListNode(3)
head2.next.next.next = ListNode(5)
head2.next.next.next.next = ListNode(6)
head2.next.next.next.next.next = ListNode(4)
head2.next.next.next.next.next.next = ListNode(7)

print("Original list - Example 2:")
current = head2
while current:
    print(current.val, end=" ")
    current = current.next
print()

reordered_head2 = oddEvenList(head2)

print("Reordered list - Example 2:")
current = reordered_head2
while current:
    print(current.val, end=" ")
    current = current.next
print()
```

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def leftShiftLinkedList(head, k):
    if not head or not head.next:
        return head

    # Step 2: Calculate the length of the linked list
    length = 0
    current = head
    while current:
        length += 1
        current = current.next

    # Step 3: Normalize the value of k
    k %= length

    # Step 4: If k is equal to 0, return the head as is
    if k == 0:
        return head

    # Step 5: Traverse to find the (k+1)-th node from the beginning
    current = head
    for _ in range(k):
        current = current.next

    # Step 6: Set the new head of the list
    new_head = current.next

    # Step 7: Traverse until the current node is the last node
    while current.next:
        current = current.next

    # Step 8: Update the current node to be its next node
    current.next = head

    # Step 9: Set the next node of the last node as the original head of the list
    # Step 10: Set the next node of the k-th node as None
    current.next.next = None

    # Step 11: Return the new head of the shifted list
    return new_head

# Example 1
head1 = ListNode(2)
head1.next = ListNode(4)
head1.next.next = ListNode(7)
head1.next.next.next = ListNode(8)
head1.next.next.next.next = ListNode(9)

print("Original list - Example 1:")
current = head1
while current:
    print(current.val, end=" ")
    current = current.next
print()

k1 = 3
shifted_head1 = leftShiftLinkedList(head1, k1)

print(f"Shifted list by {k1} nodes - Example 1:")
current = shifted_head1
while current:
    print(current.val, end=" ")
    current = current.next
print()

# Example 2
head2 = ListNode(1)
head2.next
```

You are given theÂ `head`Â of a linked list withÂ `n`Â nodes.

For each node in the list, find the value of theÂ **next greater node**. That is, for each node, find the value of the first node that is next to it and has aÂ **strictly larger**Â value than it.

Return an integer arrayÂ `answer`Â whereÂ `answer[i]`Â is the value of the next greater node of theÂ `ith`Â node (**1-indexed**). If theÂ `ith`Â node does not have a next greater node, setÂ `answer[i] = 0`.

</aside>

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def nextLargerNodes(head):
    stack = []
    answer = []
    index = 0

    while head:
        while stack and head.val > stack[-1][0]:
            value, pos = stack.pop()
            answer[pos] = head.val
        stack.append((head.val, index))
        answer.append(0)
        head = head.next
        index += 1

    while stack:
        value, pos = stack.pop()
        answer[pos] = 0

    answer.reverse()
    return answer

# Create a linked list: 2 -> 7 -> 4 -> 3 -> 5
head = ListNode(2)
head.next = ListNode(7)
head.next.next = ListNode(4)
head.next.next.next = ListNode(3)
head.next.next.next.next = ListNode(5)

# Print the original list
print("Original list:")
current = head
while current:
    print(current.val, end=" ")
    current = current.next
print()

# Find the next greater nodes
answer = nextLargerNodes(head)

# Print the next greater nodes
print("Next greater nodes:")
for val in answer:
    print(val, end=" ")
```

ðŸ’¡ **Question 8**

Given theÂ `head`Â of a linked list, we repeatedly delete consecutive sequences of nodes that sum toÂ `0`Â until there are no such sequences.

After doing so, return the head of the final linked list.Â  You may return any such answer.

(Note that in the examples below, all sequences are serializations ofÂ `ListNode`Â objects.)

**Example 1:**
Input: head = [1,2,-3,3,1]
Output: [3,1]
Note: The answer [1,2,1] would also be accepted.

Example 2:

Input: head = [1,2,3,-3,4]
Output: [1,2,4]

Example 3:

Input: head = [1,2,3,-3,-2]
Output: [1]

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def removeZeroSumSublists(head):
    dummy = ListNode(0)
    dummy.next = head
    prefix_sum = 0
    node_map = {0: dummy}
    current = head

    while current:
        prefix_sum += current.val

        if prefix_sum in node_map:
            node_to_remove = node_map[prefix_sum].next
            temp_sum = prefix_sum + node_to_remove.val

            while node_to_remove != current:
                del node_map[temp_sum]
                node_to_remove = node_to_remove.next
                temp_sum += node_to_remove.val

            node_map[prefix_sum].next = current.next

        else:
            node_map[prefix_sum] = current

        current = current.next

    return dummy.next

# Create a linked list: 1 -> 2 -> -3 -> 3 -> 1
head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(-3)
head.next.next.next = ListNode(3)
head.next.next.next.next = ListNode(1)

# Print the original list
print("Original list:")
current = head
while current:
    print(current.val, end=" ")
    current = current.next
print()

# Remove consecutive sequences of nodes with sum 0
new_head = removeZeroSumSublists(head)

# Print the final list
print("Final list:")
current = new_head
while current:
    print(current.val, end=" ")
    current = current.next
```
